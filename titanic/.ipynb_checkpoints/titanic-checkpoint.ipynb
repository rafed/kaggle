### Joining train and test so that common operations can be done together. Also filling missing data and creating new columns.

```python
train_processed = train.copy(deep=True)
test_processed = test.copy(deep=True)

train_test_data = [train_processed, test_processed] # combining train and test dataset

for dataset in train_test_data:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.')
    
    dataset['Embarked'] = dataset['Embarked'].fillna('S')

    # Age stuff
    # cut_points = [-1,0, 5, 12, 18, 35, 60, 100]
    # label_names = ["Missing", 'Infant', "Child", 'Teenager', "Young Adult", 'Adult', 'Senior']
    
    dataset["Age"] = dataset["Age"].fillna(-0.5)
    dataset.loc[(dataset['Age'] < 0) & (dataset['Title']=='Mr'), 'Age'] = 30
    dataset.loc[(dataset['Age'] < 0) & (dataset['Title']=='Miss'), 'Age'] = 18
    dataset.loc[(dataset['Age'] < 0) & (dataset['Title']=='Mrs'), 'Age'] = 30
    dataset.loc[(dataset['Age'] < 0) & (dataset['Title']=='Master'), 'Age'] = 13
    dataset.loc[(dataset['Age'] < 0) & (dataset['Title']=='Dr'), 'Age'] = 65
    
    cut_points = [-1, 5, 12, 18, 60, 100]
    label_names = ['Infant', "Child", 'Teenager', 'Adult', 'Senior']
    dataset["Age_categories"] = pd.cut(dataset["Age"],cut_points,labels=label_names)
    
    dataset['Family'] = 0
    dataset.loc[(dataset['SibSp'] > 0) | (dataset['Parch']>0), 'Family'] = 1
    
    dataset['Fare'] = dataset['Fare'].fillna(train['Fare'].median())
```

### Convert categorical values to codes

```python
# dataset['Embarked'] = dataset['Embarked'].astype("category").cat.codes
# Lets not use this, test data and train data may not get same codes

for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].astype("category").cat.codes
    
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
    
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
    
    # Age stuff
    label_names = ['Infant', "Child", 'Teenager', 'Adult', 'Senior']
    
    dataset['Age_categories'] = dataset['Age_categories'].map(
        {'Infant':0, 'Child':1, 'Teenager':2, 'Adult':3, 'Senior':4}
    )
    
    
```

### Feature selection

```python
# features_drop = ['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin']
# train = train.drop(features_drop, axis=1)
# test = test.drop(features_drop, axis=1)
# train = train.drop(['PassengerId', 'AgeBand', 'FareBand'], axis=1)

# Features: Survived, Pclass, !Name, Sex, Age, SibSp, Parch, !Ticket, Fare, !Cabin, Embarked, [Title], [Age_categories], [Family]

# selected features
features = ["Pclass", "Sex", "Embarked", "SibSp", "Parch", "Title", "Age_categories", "Family", "Fare"]

train_x = train_processed[features]
train_y = train_processed["Survived"] # use .copy() maybe??
test_x = test_processed[features]

train_x.shape, train_y.shape, test_x.shape
```

### Classification and accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier

models = []
models.append(('LR', LogisticRegression()))
models.append(('SVC', SVC()))
models.append(('LinSVC', LinearSVC()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DTR', DecisionTreeClassifier()))
models.append(('RFR', RandomForestClassifier()))
models.append(('GAU', GaussianNB()))
models.append(('NN', Perceptron()))
models.append(('SGD', SGDClassifier()))
```

```python
from sklearn import model_selection

scoring = 'accuracy' #'balanced_accuracy'
kfold = model_selection.KFold(n_splits=10, random_state=10)

model_accuracy = []

for name, model in models:
    results = model_selection.cross_val_score(model, train_x, train_y, scoring=scoring, cv=kfold)
    model_accuracy.append((name, round(results.mean()*100,2)))
    
# Print accuracy of models
model_accuracy = sorted(model_accuracy, key=lambda k: k[1], reverse=True)
for name, accuracy in model_accuracy:
    print("{:7}: {}%".format(name, accuracy))
```

### Model hyper-parameters tuning

```python
from sklearn.model_selection import GridSearchCV

model = RandomForestClassifier()
n_estim=range(100,1000,100)

## Search grid for optimal parameters
param_grid = {"n_estimators" :n_estim}
model_rf = GridSearchCV(model,param_grid = param_grid, cv=5, scoring="accuracy", n_jobs= 4, verbose = 1)
model_rf.fit(train_x,train_y)

# Best score
print(model_rf.best_score_)

#best estimator
model_rf.best_estimator_
```

### Submission

```python
model = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=None, max_features='auto', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=300, n_jobs=None,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False)
model.fit(train_x, train_y)
pred = model.predict(test_x)

submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": pred
    })

submission.to_csv('submission.csv', index=False)
```
